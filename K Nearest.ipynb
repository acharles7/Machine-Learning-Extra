```python
import tensorflow as tf
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets("MNIST/", one_hot=True)
```

```python
X = tf.placeholder(tf.float32, [None, 784])
Y = tf.placeholder(tf.float32, [784])

x_train, y_train = mnist.train.next_batch(5000)
x_test, y_test = mnist.train.next_batch(200)
```

```python
distance = tf.reduce_sum(tf.abs(X - Y), axis=1)
preds = tf.argmin(distance, 0)
```

```python
accuracy = 0

with tf.Session() as sess:
    init = tf.global_variables_initializer()
    sess.run(init)
    
    for i in range(len(x_test)):
        index = sess.run(preds, feed_dict={X:x_train, Y:x_test[i, :]})
        print("Prediction:", np.argmax(y_train[index]), "True: ",np.argmax(y_test[i]))
        
        if np.argmax(y_train[index]) == np.argmax(y_test[i]):
            accuracy += 1 / len(x_test)
    print("Accuracy: ",accuracy)
        
        
        
```
